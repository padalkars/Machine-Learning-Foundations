```python
##Pearsons Correlation
def pearson(x,y):
    sd_x = np.sqrt(np.sum((x-np.mean(x))**2)/len(x))
    sd_y = np.sqrt(np.sum((y-np.mean(y))**2)/len(y))
    return (np.sum(x-np.mean(x)) * np.sum(y-np.mean(y)))/(sd_x * sd_y)
```

```python
#Function to find correlation between a movie and rest all the movies other than the movie
def find_correlated(movie, M, num):
    corr = []
    for title in M.columns:
        #A movie will have a high correlation with itself
        if(title == movie):
            continue
        val = pearson(M[title], M[movie])
        if(np.isnan(val)):
            continue
        
        corr.append((title, val))
    corr.sort(key=lambda tup: tup[1], reverse = True) 
    
    if(num>0):
        #Return the top 'num' highly correlated titles
        return corr[:num]
    if(num==0):
        return corr[num]
    if(num<0):
        #Return the top 'num' less correlated titles
        return corr[num:]
```

```python
find_correlated('Mission: Impossible II (2000)', M, 10)
```

```python
find_correlated('Clerks (1994)', M, 10)
```

```python
find_correlated('Clerks (1994)', M, -10)
```
